## Sorting

```python
unsorted = np.array([21,5,8,3,7,9,4,23])
print("unsorted : ", unsorted)
```

```python
sorted = np.sort(unsorted)
print(sorted)
print(unsorted)
```

```python
# 2D Arrays Sorting 

arr_2d = np.array([
    [34, 11, 56],
    [21, 90, 12],
    [67, 45, 89]
])

# Sort each row (axis=1)
sorted_rows = np.sort(arr_2d, axis=1)
print("Row-wise sorting (axis=1):\n", sorted_rows)

# Sort each column (axis=0)
sorted_columns = np.sort(arr_2d, axis=0)
print("Column-wise sorting (axis=0):\n", sorted_columns)
```

 ## Filtering 

```python


arr = np.array([10, 25, 37, 14, 8, 42, 19])
print("Original array is : ", arr);
# Filter values greater than 20
filtered = arr[arr > 20]
print("\nFiltered values greater than 20:", filtered)


# 2D
arr_2d = np.array([
    [5, 12, 19],
    [8, 25, 30],
    [15, 7, 22]
])

# Filter values greater than 10
mask = arr_2d > 10 
print("\nMask looks like : ", mask)
filtered_2d = arr_2d[mask]
print("\nFiltered values greater than 10:", filtered_2d)


# using where 
mask2 = np.where(arr > 15)
print("\nMask of .where : (return indexes) ",mask2)

# Modify array using where condition
updated_arr = np.where(arr>15 , arr*2, arr*1)  # condition , if true , else 
print("\nUpdated array after where condition updates: ", updated_arr)


# check original array 
print("\nOriginal array in the end: ", arr)
```

## Concatenate

# Advcance operations with buisness examples 

```python
import matplotlib.pyplot as plt
restaurants = np.array([
    [101, 4.5, 120, 2, 35],  # Restaurant 1
    [102, 3.8, 85, 1, 20],   # Restaurant 2
    [103, 4.2, 200, 3, 50],  # Restaurant 3
    [104, 4.0, 150, 2, 40],  # Restaurant 4
    [105, 3.5, 60, 1, 25]    # Restaurant 5
])



labels = ["ratings", "No of reviews" , "Price range", "No of dishes"]
data= restaurants[: , 1:]
print("Data is : \n",data)

sum_data = np.sum(data,axis=0)
print("\n\nSummed Data : " , sum_data)



# Extracting data
restaurant_ids = restaurants[:, 0].astype(int)  # Restaurant IDs
ratings = restaurants[:, 1]                     # Ratings
reviews = restaurants[:, 2]                     # Number of reviews
price_ranges = restaurants[:, 3]                # Price range (1=Cheap, 2=Moderate, 3=Expensive)
dishes = restaurants[:, 4]                      # Dishes available

# 1. Bar Chart - Number of Reviews per Restaurant
plt.figure(figsize=(10, 5))
plt.bar(restaurant_ids, reviews, color='skyblue', edgecolor='black')
plt.xlabel("Restaurant ID")
plt.ylabel("Number of Reviews")
plt.title("Number of Reviews per Restaurant")

plt.show()




# 2. Scatter Plot - Ratings vs. Number of Reviews
plt.figure(figsize=(10, 5))
plt.scatter(reviews, ratings, color='red', marker='o')
plt.xlabel("Number of Reviews")
plt.ylabel("Ratings")
plt.title("Ratings vs. Number of Reviews")
plt.grid(True)
plt.show()

# 3. Pie Chart - Price Range Distribution
labels = ["Cheap", "Moderate", "Expensive"]
price_counts = [np.sum(price_ranges == 1), np.sum(price_ranges == 2), np.sum(price_ranges == 3)]

plt.figure(figsize=(7, 7))
plt.pie(price_counts, labels=labels, autopct='%1.1f%%', colors=['lightgreen', 'orange', 'purple'])
plt.title("Distribution of Price Ranges")
plt.show()
```
